- Decoding the values
- We will measure quality of compatibility of personality of medcal students as a sub-goal in article

```python
# From AboliEEEEEE
personality = {'الف':'اجتماعی','و':'واقع گرا','ج':'جستجوگر','م':'متهورانه','ه':'هنری','ق':'قراردادی'}
hamkhani = {'اجتماعی':'پایین','واقع گرا':'متوسط','جستجوگر':'کامل',
            'متهورانه':'عدم همخوانی','هنری':'متوسط','قراردادی':'پایین'}
faliat_job = {'و':'فنی', 'الف':'فرهنگی', 'ج':'علمی', 'م':'اجتماعی', 'ه':'هنری', 'ق':'عمومی'}

for i in range(len(sep)):
    sep.loc[i,'تیپ غالب'] = personality[sep.loc[i,'1تیپ']]

for i in range(len(sep)):
    sep.loc[i,'همخوانی'] = hamkhani[sep.loc[i,'تیپ غالب']]
sep
```

```python
# Initialing the columns
sep['1فعالیت'] = 0
sep['2فعالیت'] = 0
sep['3فعالیت'] = 0
sep['4فعالیت'] = 0

faliats = []
for row in range(len(sep)):
    m = max(sep.iloc[row,11:17])
    mask = sep.iloc[row,11:17][sep.iloc[row,11:17] == m]
    values = list(mask.index.values)
    faliats.append(values)

for i in range(len(faliats)):
    l = len(faliats[i])
    if l > 4: # bazi ha hame ro zadan
        faliats[i] = faliats[i][:-(l-4)]
    tedad_0 = 4-l
    for _ in range(tedad_0):
        faliats[i].append(0)

# sep[['فعالیت1','فعالیت2','فعالیت3','فعالیت4']] = faliats
sep[['1فعالیت', '2فعالیت', '3فعالیت', '4فعالیت']] = faliats

for row in range(len(sep)):
    if sep['1فعالیت'][row] != 0:
        sep.loc[row,'فعالیت1'] = faliat_job[sep['1فعالیت'][row]]
    if sep['2فعالیت'][row] != 0:
        sep.loc[row,'فعالیت2'] = faliat_job[sep['2فعالیت'][row]]
    if sep['3فعالیت'][row] != 0:
        sep.loc[row,'فعالیت3'] = faliat_job[sep['3فعالیت'][row]]
    if sep['4فعالیت'][row] != 0:
        sep.loc[row,'فعالیت4'] = faliat_job[sep['4فعالیت'][row]]
sep = sep.fillna(0)
sep.drop(sep.columns[[-8,-7,-6,-5]],axis=1, inplace=True)
```

The Values in columns `[27,28,29,30,31,32]` are related to the persons' job

```python
sep['1کار'] = 0
sep['2کار'] = 0
sep['3کار'] = 0
sep['4کار'] = 0
faliats = []

for row in range(len(sep)):
    m = max(sep.iloc[row,23:29])
    mask = sep.iloc[row,23:29][sep.iloc[row,23:29] == m]
    values = list(mask.index.values)
    faliats.append(values)

for i in range(len(faliats)):
    l = len(faliats[i])
    if l > 4: # bazi ha hame ro zadan
        faliats[i] = faliats[i][:-(l-4)]
    tedad_0 = 4-l
    for _ in range(tedad_0):
        faliats[i].append(0)

# sep[['فعالیت1','فعالیت2','فعالیت3','فعالیت4']] = faliats
sep[['1کار', '2کار', '3کار', '4کار']] = faliats

for row in range(len(sep)):
    if sep['1کار'][row] != 0:
        sep.loc[row,'کار1'] = faliat_job[sep['1کار'][row]]
    if sep['2کار'][row] != 0:
        sep.loc[row,'کار2'] = faliat_job[sep['2کار'][row]]
    if sep['3کار'][row] != 0:
        sep.loc[row,'کار3'] = faliat_job[sep['3کار'][row]]
    if sep['4کار'][row] != 0:
        sep.loc[row,'کار4'] = faliat_job[sep['4کار'][row]]
sep = sep.fillna(0)
sep.drop(sep.columns[[-8,-7,-6,-5]],axis=1, inplace=True)
```
